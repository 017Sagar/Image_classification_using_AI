```python
d = {'Yes':1,'No':0}
data["HeartDisease"]=data["HeartDisease"].map(d).fillna(data["HeartDisease"])

data["Smoking"]=data["Smoking"].map(d).fillna(data["Smoking"])

data["AlcoholDrinking"]=data["AlcoholDrinking"].map(d).fillna(data["AlcoholDrinking"])

data["Stroke"]=data["Stroke"].map(d).fillna(data["Stroke"])

data["DiffWalking"]=data["DiffWalking"].map(d).fillna(data["DiffWalking"])

data["Diabetic"]=data["Diabetic"].map(d).fillna(data["Diabetic"])

data["PhysicalActivity"]=data["PhysicalActivity"].map(d).fillna(data["PhysicalActivity"])

data["Asthma"]=data["Asthma"].map(d).fillna(data["Asthma"])

data["KidneyDisease"]=data["KidneyDisease"].map(d).fillna(data["KidneyDisease"])

data["SkinCancer"]=data["SkinCancer"].map(d).fillna(data["SkinCancer"])


```

```python
data
d = {'male':1,'female':0}
data["Sex"]=data["Sex"].map(d).fillna(data["Sex"])



```

```python
d = {'Male':1,'Female':0}
data["Sex"]=data["Sex"].map(d).fillna(data["Sex"])
```

```python
f={'White':0,'Hispanic':1}
data["Race"]=data["Race"].map(f).fillna(data["Race"])
```

```python
g={'Very good':0,'Fair':1,'Good':2}
data["GenHealth"]=data["GenHealth"].map(g).fillna(data["GenHealth"])
```

```python
data=data.drop(['AgeCategory','Race','Diabetic','GenHealth'],axis=1)
```

```python
X=data.drop('HeartDisease',axis=1)
y=data.HeartDisease
```

```python
y = data.HeartDisease.ravel()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.1)
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train = sc.fit_transform(X_train)
X_test= sc.transform(X_test)
from sklearn.svm import SVC
model=SVC()
model.fit(X_train, y_train)
model.score(X_test, y_test)
```
